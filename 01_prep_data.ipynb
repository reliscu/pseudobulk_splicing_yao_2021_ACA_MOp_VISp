## Gene counts

```python
adata_ACA = ad.read_h5ad("ACA/yao_2021_ACA_STAR_gene_counts.h5ad")
adata_MOp = ad.read_h5ad("MOp/yao_2021_MOp_STAR_gene_counts.h5ad")
adata_VISp = ad.read_h5ad("VISp/yao_2021_VISp_STAR_gene_counts.h5ad")
```

```python
adata_ACA.obs.rename(columns={'external_donor_name_label': 'donor_id'}, inplace=True)
adata_VISp.obs.rename(columns={'external_donor_name_label': 'donor_id'}, inplace=True)
```

```python
adata = ad.concat(
    {"ACA": adata_ACA, "MOp": adata_MOp, "VISp": adata_VISp},   # keys become batch labels
    axis=0,                                 # concat observations (cells)
    label="region",                          # adds adata.obs["region"]
    join="inner",                           
    merge="same",                            # keep only identical .uns entries
    index_unique="-",                        # make obs_names unique like "cell-ACA"
)
```

```python
adata.obs.value_counts("subclass_label")
```

```python
for col in adata.obs.columns:
    if adata.obs[col].dtype == object:
        adata.obs[col] = adata.obs[col].astype(str)
```

```python
adata.write(f"{outdir}/data/yao_2021_ACA_MOp_VISp_STAR_gene_counts.h5ad")
```

```python
# Also get data into format compatbile with pseudobulk function

df = pd.DataFrame(
    adata.X.T.toarray(),
    columns=adata.obs_names,
    index=adata.var_names
)

df.index.name = "Gene"

df.to_csv(f"{outdir}/data/yao_2021_ACA_MOp_VISp_STAR_gene_counts.csv")
adata.obs.to_csv(f"{outdir}/data/yao_2021_ACA_MOp_VISp_STAR_sampleinfo.csv")
```
